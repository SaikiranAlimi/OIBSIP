# **Iris Flower Classification**

# *importing the libraries*

```python
#importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# *Loading the dataset*

```python
#read the iris.csv data file
df=pd.read_csv("Iris.csv")
df.head()
```

```python
df = df.drop(columns="Id")
```

# *Data Visualization*

```python
sns.pairplot(df, hue='Species', diag_kind='hist')
plt.show()
```

# *Correlation Matrix*

```python
#compute  the correlation matrix
df.corr()
```

# *creating the correlation heatmap*

```python
#displaying the correlation matrix using the heatmap
corr = df.corr()
fig, ax = plt.subplots(figsize=(5, 4))
sns.heatmap(corr, annot=True, ax=ax, cmap='Spectral')
```

# *Model Training*

```python
#Splitting the datasets into training and testing sets for training and testing the model
from sklearn.model_selection import train_test_split
X = df.drop(columns=['Species'])
Y = df['Species']
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.40)
```

```python
#LOGISTIC REGRESSION MODEL


from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x_train, y_train)
print("Accuracy (Logistic Regression): ", model1.score(x_test, y_test) * 100)
```
